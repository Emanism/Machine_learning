# Calculate the variance and standard deviation for numerical

```python
#importing this again because it stop working
import pandas as pd

# Reload the dataset
df = pd.read_csv(r'C:/Users/moin/Downloads/ML Files/Car_sales.csv')

# Strip any leading/trailing whitespace from column names
df.columns = df.columns.str.strip()

# Print the DataFrame column names to confirm
print("Column Names in DataFrame:", df.columns.tolist())
```

```python
# 5. Variance and Standard Deviation
# Calculate variance and standard deviation for the selected columns
var_std_df = df[['Sales_in_thousands', '__year_resale_value', 'Price_in_thousands', 'Engine_size', 'Horsepower']].agg(['var', 'std'])
print(var_std_df)
```

# For the same columns Calculate IQR and identify outliers in these columns

```python
# 6. Interquartile Range (IQR) and Outliers
# Calculate the IQR for the selected numeric columns

# Select only numeric columns for quantile calculation
numeric_df = df.select_dtypes(include=[np.number])

# Calculate the 1st and 3rd quartiles (Q1 and Q3)
Q1 = numeric_df.quantile(0.25)
Q3 = numeric_df.quantile(0.75)

# Calculate the Interquartile Range (IQR)
IQR = Q3 - Q1

print("IQR for numeric columns:\n", IQR)

outliers = numeric_df[(numeric_df < (Q1 - 1.5 * IQR)) | (numeric_df > (Q3 + 1.5 * IQR))]

print("Outliers:\n", outliers)
```

# Make Histograms for all Numerical columns and identify the distribution of type by calculating skewness;

```python
# 7. Histograms and Skewness

# Histograms for all numerical columns
numeric_df.hist(figsize=(12, 10))
plt.tight_layout()
plt.show()

# Skewness for numerical columns
skewness = numeric_df.skew()
print("Skewness for numerical columns:\n", skewness)
```
